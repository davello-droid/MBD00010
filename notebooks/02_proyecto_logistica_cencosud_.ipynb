## BLOQUE 4 — Funciones auxiliares

```python
def leer_csv_seguro(path_obj):
    """
    Intenta leer un CSV con varias combinaciones comunes de encoding y separador.
    Prioriza ';' porque los archivos generados por el notebook de preparación usan ese separador.
    """
    intentos = [
        {"sep": ";", "encoding": "utf-8-sig"},
        {"sep": ";", "encoding": "utf-8"},
        {"sep": ";", "encoding": "latin-1"},
        {"sep": ",", "encoding": "utf-8-sig"},
        {"sep": ",", "encoding": "utf-8"},
        {"sep": ",", "encoding": "latin-1"},
    ]
    ultimo_error = None
    for intento in intentos:
        try:
            df_tmp = pd.read_csv(path_obj, sep=intento["sep"], encoding=intento["encoding"], low_memory=False)
            if df_tmp.shape[1] > 1:
                print(f"[OK] Lectura exitosa: {path_obj.name} | sep='{intento['sep']}' | encoding='{intento['encoding']}'")
                return df_tmp
        except Exception as e:
            ultimo_error = e
    raise ValueError(f"No fue posible leer el archivo: {path_obj}. Último error detectado: {ultimo_error}")


def estandarizar_columnas(df):
    df = df.copy()
    df.columns = [str(c).strip().lower() for c in df.columns]
    return df


def detectar_columna_fecha(df):
    candidatos = ["fectra", "fecha", "fecha_documento", "fecha_registro", "fec_tra", "fechadoc"]
    return next((c for c in candidatos if c in df.columns), None)


def detectar_columna_codigo_arancel(df):
    candidatos = ["codigo_arancel", "aranc_nac", "codigo_hs", "partida", "hs", "codigo"]
    return next((c for c in candidatos if c in df.columns), None)


def parsear_fecha_robusta(serie):
    """
    Convierte fechas almacenadas como enteros o texto en distintos formatos observados en los datos.
    """
    s = pd.Series(serie, index=serie.index)
    out = pd.Series(pd.NaT, index=serie.index, dtype="datetime64[ns]")

    try:
        direct = pd.to_datetime(s, errors="coerce", dayfirst=True, format="mixed")
    except Exception:
        direct = pd.to_datetime(s, errors="coerce", dayfirst=True)

    out.loc[direct.notna()] = direct.loc[direct.notna()]

    mask = out.isna()
    if mask.any():
        parsed_individual = s.loc[mask].apply(lambda x: pd.to_datetime(x, errors="coerce", dayfirst=True))
        out.loc[parsed_individual.notna().index] = out.loc[parsed_individual.notna().index]
        idx_ok = parsed_individual.notna()
        if idx_ok.any():
            out.loc[parsed_individual.index[idx_ok]] = parsed_individual.loc[idx_ok].values

    mask = out.isna()
    if mask.any():
        s_txt = s.loc[mask].astype(str).str.strip()
        s_num = s_txt.str.replace(r"[^0-9]", "", regex=True)

        excel_mask = s_num.str.fullmatch(r"\d{5}")
        if excel_mask.any():
            excel_vals = pd.to_numeric(s_num.loc[excel_mask], errors="coerce")
            excel_dates = pd.to_datetime(excel_vals, unit="D", origin="1899-12-30", errors="coerce")
            out.loc[excel_dates.index] = excel_dates.values

        mask = out.isna()
        if mask.any():
            s_num = s.loc[mask].astype(str).str.strip().str.replace(r"[^0-9]", "", regex=True)
            s_num = s_num.where(~s_num.str.len().eq(7), "0" + s_num)

            for fmt in ["%d%m%Y", "%Y%m%d", "%d%m%y", "%y%m%d"]:
                parsed = pd.to_datetime(s_num, format=fmt, errors="coerce")
                idx_ok = parsed.notna()
                if idx_ok.any():
                    out.loc[parsed.index[idx_ok]] = parsed.loc[idx_ok].values

    return out

def reconstruir_hs4(df, col_hs4="hs4", col_codigo_arancel=None):
    """
    Reconstruye hs4 de forma robusta.
    Prioridad:
    1) hs4 existente
    2) codigo_arancel
    """
    df = df.copy()
    base = df[col_hs4].astype(str) if col_hs4 in df.columns else None
    if col_codigo_arancel and col_codigo_arancel in df.columns:
        base_alt = df[col_codigo_arancel].astype(str)
        if base is None:
            base = base_alt
        else:
            vacio = base.astype(str).str.strip().isin(["", "nan", "none"])
            base = pd.Series(np.where(vacio, base_alt, base), index=df.index)
    if base is None:
        df["hs4_limpio"] = np.nan
        return df
    df["hs4_limpio"] = pd.Series(base, index=df.index).astype(str).str.strip().str.replace(r"[^0-9]", "", regex=True).str[:4]
    return df


def asegurar_numericas(df, cols):
    df = df.copy()
    faltantes = [c for c in cols if c not in df.columns]
    if faltantes:
        raise ValueError(f"Faltan columnas numéricas esperadas: {faltantes}")
    for c in cols:
        df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)
    return df


_BASE_FEATURES = ["t", "mes", "anio", "lag_1", "lag_2", "lag_3", "rolling_3", "rolling_6"]


def crear_features_temporales(df, target_col):
    """
    Crea variables simples de series de tiempo para modelos supervisados.
    """
    d = df.copy().sort_values("fecha_mes").reset_index(drop=True)
    d["t"] = np.arange(len(d))
    d["mes"] = d["fecha_mes"].dt.month
    d["anio"] = d["fecha_mes"].dt.year
    for i in [1, 2, 3]:
        d[f"lag_{i}"] = d[target_col].shift(i)
    d["rolling_3"] = d[target_col].shift(1).rolling(3).mean()
    d["rolling_6"] = d[target_col].shift(1).rolling(6).mean()
    return d


def metricas_regresion(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    try:
        mape = np.mean(np.abs((y_true - y_pred) / np.where(y_true == 0, 1, y_true))) * 100
    except Exception:
        mape = np.nan
    return {"mae": mae, "rmse": rmse, "mape": mape, "r2": r2_score(y_true, y_pred) if len(y_true) > 1 else np.nan}


def _split_supervisado(d, target_col, nombre_modelo, features):
    d = d.dropna().copy()
    if len(d) < 12:
        return None, None, None, None, f"No hay suficientes datos para {nombre_modelo}"
    test_size = max(6, int(len(d) * 0.2))
    if len(d) <= test_size:
        return None, None, None, None, f"Datos insuficientes para particionar en {nombre_modelo}"
    X, y = d[features], d[target_col]
    return d, X.iloc[:-test_size], X.iloc[-test_size:], y.iloc[:-test_size], y.iloc[-test_size:], None


def entrenar_y_evaluar_supervisado(df_feat, target_col, nombre_modelo, modelo):
    split = _split_supervisado(df_feat, target_col, nombre_modelo, _BASE_FEATURES)
    if split[-1] is not None:
        return None, None, split[-1]
    d, X_train, X_test, y_train, y_test, _ = split
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)
    mets = metricas_regresion(y_test, y_pred)
    mets["modelo"] = nombre_modelo
    pred_df = pd.DataFrame({"fecha_mes": d.iloc[-len(y_test):]["fecha_mes"].values, "y_real": y_test.values, "y_pred": y_pred})
    return mets, pred_df, None


def _extender_y_recalcular(futuro, target_col, exog_cols=None):
    futuro["t"] = np.arange(len(futuro))
    futuro["mes"] = futuro["fecha_mes"].dt.month
    futuro["anio"] = futuro["fecha_mes"].dt.year
    for i in [1, 2, 3]:
        futuro[f"lag_{i}"] = futuro[target_col].shift(i)
    futuro["rolling_3"] = futuro[target_col].shift(1).rolling(3).mean()
    futuro["rolling_6"] = futuro[target_col].shift(1).rolling(6).mean()
    for c in exog_cols or []:
        if c in futuro.columns:
            futuro[f"{c}_lag1"] = futuro[c].shift(1)
            futuro[f"{c}_lag2"] = futuro[c].shift(2)
    return futuro


def forecast_supervisado(df_feat, target_col, modelo):
    d = df_feat.copy().sort_values("fecha_mes").reset_index(drop=True)
    d_model = d.dropna().copy()
    if len(d_model) < 12:
        return None
    modelo.fit(d_model[_BASE_FEATURES], d_model[target_col])
    futuro, ultima_fecha = d.copy(), d["fecha_mes"].max()
    for _ in range(HORIZONTE_MESES):
        nueva_fecha = ultima_fecha + pd.offsets.MonthBegin(1)
        futuro = pd.concat([futuro, pd.DataFrame([{"fecha_mes": nueva_fecha, target_col: np.nan}])], ignore_index=True)
        futuro = _extender_y_recalcular(futuro, target_col)
        futuro.loc[futuro.index[-1], target_col] = modelo.predict(futuro.iloc[[-1]][_BASE_FEATURES])[0]
        ultima_fecha = nueva_fecha
    return futuro.tail(HORIZONTE_MESES)[["fecha_mes", target_col]].rename(columns={target_col: "forecast"})


def _train_test_ts(df_ts, target_col, min_len=18, extra_cols=None):
    cols = ["fecha_mes", target_col] + (extra_cols or [])
    d = df_ts[cols].dropna().copy().sort_values("fecha_mes")
    if len(d) < min_len:
        return None, None, None, None, None
    test_size = max(6, int(len(d) * 0.2))
    return d, d.iloc[:-test_size], d.iloc[-test_size:], test_size, d["fecha_mes"].iloc[-test_size:].values


def entrenar_y_evaluar_arima(df_ts, target_col):
    d, train, test, test_size, fechas_test = _train_test_ts(df_ts, target_col)
    if d is None:
        return None, None, "No hay suficientes datos para ARIMA"
    try:
        ajuste = ARIMA(train[target_col].astype(float).values, order=(1, 1, 1)).fit()
        pred = ajuste.forecast(steps=test_size)
        mets = metricas_regresion(test[target_col].values, pred)
        mets["modelo"] = "ARIMA"
        pred_df = pd.DataFrame({"fecha_mes": fechas_test, "y_real": test[target_col].values, "y_pred": pred})
        return mets, pred_df, None
    except Exception as e:
        return None, None, str(e)


def forecast_arima(df_ts, target_col):
    d, *_ = _train_test_ts(df_ts, target_col)
    if d is None:
        return None
    try:
        pred = ARIMA(d[target_col].astype(float).values, order=(1, 1, 1)).fit().forecast(steps=HORIZONTE_MESES)
        fechas_futuras = pd.date_range(start=d["fecha_mes"].max() + pd.offsets.MonthBegin(1), periods=HORIZONTE_MESES, freq="MS")
        return pd.DataFrame({"fecha_mes": fechas_futuras, "forecast": pred})
    except Exception:
        return None


def guardar_tabla(df, nombre_archivo):
    ruta = TABLES_DIR / nombre_archivo
    df.to_csv(ruta, index=False, encoding="utf-8-sig")
    print(f"[OK] Tabla guardada: {ruta}")


def guardar_figura(nombre_archivo):
    ruta = GRAPHICS_DIR / nombre_archivo
    plt.savefig(ruta, bbox_inches="tight", dpi=150)
    print(f"[OK] Gráfico guardado: {ruta}")


MESES_ES = {
    "ene": 1, "enero": 1,
    "feb": 2, "febrero": 2,
    "mar": 3, "marzo": 3,
    "abr": 4, "abril": 4,
    "may": 5, "mayo": 5,
    "jun": 6, "junio": 6,
    "jul": 7, "julio": 7,
    "ago": 8, "agosto": 8,
    "sep": 9, "sept": 9, "septiembre": 9,
    "oct": 10, "octubre": 10,
    "nov": 11, "noviembre": 11,
    "dic": 12, "diciembre": 12
}

def _normalizar_texto_mes(texto):
    import unicodedata
    txt = str(texto).strip().lower()
    txt = "".join(ch for ch in unicodedata.normalize("NFKD", txt) if not unicodedata.combining(ch))
    txt = re.sub(r"[^a-z0-9]", "", txt)
    return txt


def _decimal_coma_a_float(serie):
    return pd.to_numeric(serie.astype(str).str.replace(".", "", regex=False).str.replace(",", ".", regex=False), errors="coerce")


def _parsear_fecha_es(fecha_str):
    if pd.isna(fecha_str):
        return pd.NaT

    txt = str(fecha_str).strip().lower()
    txt = re.sub(r"\s+", " ", txt)
    m = re.match(r"^(\d{1,2})[\.\-/ ]+([a-zñáéíóú]+)[\.\-/ ]+(\d{4})$", txt)
    if not m:
        return pd.NaT

    dia, mes_txt, anio = m.groups()
    mes_key = _normalizar_texto_mes(mes_txt)
    mes = MESES_ES.get(mes_key)
    if mes is None:
        mes = MESES_ES.get(mes_key[:4], MESES_ES.get(mes_key[:3]))

    if mes is None:
        return pd.NaT

    return pd.Timestamp(year=int(anio), month=int(mes), day=int(dia))

def _parsear_mes_es(mes_str):
    if pd.isna(mes_str):
        return pd.NaT

    txt = str(mes_str).strip().lower()
    txt = re.sub(r"\s+", " ", txt)
    m = re.match(r"^([a-zñáéíóú]+)[\.\-/ ]+(\d{4})$", txt)
    if not m:
        return pd.NaT

    mes_txt, anio = m.groups()
    mes_key = _normalizar_texto_mes(mes_txt)
    mes = MESES_ES.get(mes_key)
    if mes is None:
        mes = MESES_ES.get(mes_key[:4], MESES_ES.get(mes_key[:3]))

    if mes is None:
        return pd.NaT

    return pd.Timestamp(year=int(anio), month=int(mes), day=1)

def _cargar_bcch_diario(path_csv, nombre_columna):
    ultimo_error = None
    df = None

    for enc in ["utf-8", "utf-8-sig", "latin-1", "cp1252"]:
        try:
            df = pd.read_csv(path_csv, sep=",", encoding=enc)
            break
        except Exception as e:
            ultimo_error = e

    if df is None:
        raise ultimo_error

    df = estandarizar_columnas(df)
    col_fecha = next((c for c in df.columns if c in ["dia", "fecha"]), None)
    col_valor = next((c for c in df.columns if c == "valor"), None)

    if col_fecha is None or col_valor is None:
        raise ValueError(f"El archivo {path_csv.name} no contiene las columnas esperadas de fecha/valor.")

    df["fecha"] = df[col_fecha].apply(_parsear_fecha_es)
    df[nombre_columna] = _decimal_coma_a_float(df[col_valor])
    df["fecha_mes"] = df["fecha"].dt.to_period("M").dt.to_timestamp()

    return (
        df.groupby("fecha_mes", as_index=False)[nombre_columna]
        .mean()
        .sort_values("fecha_mes")
        .reset_index(drop=True)
    )

def cargar_bcch_dolar_observado(path_csv):
    return _cargar_bcch_diario(path_csv, "bcch_dolar_observado")


def cargar_bcch_precio_cobre(path_csv):
    return _cargar_bcch_diario(path_csv, "bcch_precio_cobre")


def cargar_bcch_ipc_mensual(path_csv):
    ultimo_error, df = None, None
    for enc in ["latin-1", "cp1252", "utf-8", "utf-8-sig"]:
        try:
            df = pd.read_csv(path_csv, sep=",", encoding=enc)
            break
        except Exception as e:
            ultimo_error = e
    if df is None:
        raise ultimo_error

    df = estandarizar_columnas(df)
    col_desc = next((c for c in df.columns if "descrip" in c), df.columns[1])
    texto = df[col_desc].astype(str).str.lower()
    fila = df[texto.str.contains("ipc", na=False) & ~texto.str.contains("sae", na=False)].copy()
    if fila.empty:
        fila = df.iloc[[0]].copy()

    cols_mes = []
    for c in fila.columns:
        partes = re.split(r"[\.\-/ ]+", str(c).strip().lower())
        if len(partes) == 2 and partes[1].isdigit() and len(partes[1]) == 4:
            cols_mes.append(c)

    ipc_long = fila.melt(
        id_vars=[c for c in fila.columns if c not in cols_mes],
        value_vars=cols_mes,
        var_name="periodo",
        value_name="bcch_ipc_mensual"
    )
    ipc_long["fecha_mes"] = ipc_long["periodo"].apply(_parsear_mes_es)
    ipc_long["bcch_ipc_mensual"] = _decimal_coma_a_float(ipc_long["bcch_ipc_mensual"])

    return (
        ipc_long[["fecha_mes", "bcch_ipc_mensual"]]
        .dropna()
        .sort_values("fecha_mes")
        .reset_index(drop=True)
    )

def integrar_variables_externas_bcch(monthly_total, data_dir):
    base = monthly_total.copy().sort_values("fecha_mes").reset_index(drop=True)
    fuentes = {
        "bcch_dolar_observado.csv": cargar_bcch_dolar_observado,
        "bcch_ipc_mensual.csv": cargar_bcch_ipc_mensual,
        "bcch_precio_cobre.csv": cargar_bcch_precio_cobre,
    }
    for nombre, loader in fuentes.items():
        ruta = data_dir / nombre
        if ruta.exists():
            base = base.merge(loader(ruta), on="fecha_mes", how="left")
        else:
            print(f"[WARNING] No se encontró: {ruta}")
    cols_ext = [c for c in ["bcch_dolar_observado", "bcch_ipc_mensual", "bcch_precio_cobre"] if c in base.columns]
    if cols_ext:
        base[cols_ext] = base[cols_ext].sort_index().ffill()
    return base


def crear_features_temporales_ext(df, target_col, exog_cols=None):
    d = crear_features_temporales(df, target_col)
    for c in exog_cols or []:
        if c in d.columns:
            d[f"{c}_lag1"] = d[c].shift(1)
            d[f"{c}_lag2"] = d[c].shift(2)
    return d


def _features_ext(df, exog_cols=None):
    features = list(_BASE_FEATURES)
    for c in exog_cols or []:
        for suff in ["_lag1", "_lag2"]:
            colx = f"{c}{suff}"
            if colx in df.columns:
                features.append(colx)
    return features


def entrenar_y_evaluar_supervisado_ext(df_feat, target_col, nombre_modelo, modelo, exog_cols=None):
    features = _features_ext(df_feat, exog_cols)
    split = _split_supervisado(df_feat, target_col, nombre_modelo, features)
    if split[-1] is not None:
        return None, None, split[-1]
    d, X_train, X_test, y_train, y_test, _ = split
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)
    mets = metricas_regresion(y_test, y_pred)
    mets["modelo"] = nombre_modelo
    pred_df = pd.DataFrame({"fecha_mes": d.iloc[-len(y_test):]["fecha_mes"].values, "y_real": y_test.values, "y_pred": y_pred})
    return mets, pred_df, None


def forecast_supervisado_ext(df_feat, target_col, modelo, exog_cols=None):
    d = df_feat.copy().sort_values("fecha_mes").reset_index(drop=True)
    d_model = d.dropna().copy()
    if len(d_model) < 12:
        return None
    features = _features_ext(d_model, exog_cols)
    modelo.fit(d_model[features], d_model[target_col])
    futuro, ultima_fecha = d.copy(), d["fecha_mes"].max()
    ult_exog = {c: (futuro[c].dropna().iloc[-1] if c in futuro.columns and futuro[c].dropna().shape[0] > 0 else np.nan) for c in (exog_cols or [])}
    for _ in range(HORIZONTE_MESES):
        nueva_fecha = ultima_fecha + pd.offsets.MonthBegin(1)
        nueva_fila = {"fecha_mes": nueva_fecha, target_col: np.nan, **ult_exog}
        futuro = pd.concat([futuro, pd.DataFrame([nueva_fila])], ignore_index=True)
        futuro = _extender_y_recalcular(futuro, target_col, exog_cols)
        futuro.loc[futuro.index[-1], target_col] = modelo.predict(futuro.iloc[[-1]][features])[0]
        ultima_fecha = nueva_fecha
    return futuro.tail(HORIZONTE_MESES)[["fecha_mes", target_col]].rename(columns={target_col: "forecast"})


def entrenar_y_evaluar_exp_smoothing(df_ts, target_col):
    d, train, test, test_size, fechas_test = _train_test_ts(df_ts, target_col)
    if d is None:
        return None, None, "No hay suficientes datos para Exponential Smoothing"
    try:
        pred = ExponentialSmoothing(train[target_col].astype(float).values, trend="add", seasonal=None).fit(optimized=True).forecast(test_size)
        mets = metricas_regresion(test[target_col].values, pred)
        mets["modelo"] = "Exponential Smoothing"
        pred_df = pd.DataFrame({"fecha_mes": fechas_test, "y_real": test[target_col].values, "y_pred": pred})
        return mets, pred_df, None
    except Exception as e:
        return None, None, str(e)


def forecast_exp_smoothing(df_ts, target_col):
    d, *_ = _train_test_ts(df_ts, target_col)
    if d is None:
        return None
    try:
        pred = ExponentialSmoothing(d[target_col].astype(float).values, trend="add", seasonal=None).fit(optimized=True).forecast(HORIZONTE_MESES)
        fechas_futuras = pd.date_range(start=d["fecha_mes"].max() + pd.offsets.MonthBegin(1), periods=HORIZONTE_MESES, freq="MS")
        return pd.DataFrame({"fecha_mes": fechas_futuras, "forecast": pred})
    except Exception:
        return None


def entrenar_y_evaluar_sarimax(df_ts, target_col, exog_cols):
    if len(exog_cols) == 0:
        return None, None, "No hay suficientes datos o exógenas para SARIMAX"
    d, train, test, test_size, fechas_test = _train_test_ts(df_ts, target_col, extra_cols=exog_cols)
    if d is None:
        return None, None, "No hay suficientes datos o exógenas para SARIMAX"
    try:
        ajuste = SARIMAX(
            train[target_col].astype(float).values,
            exog=train[exog_cols].astype(float).values,
            order=(1, 1, 1),
            enforce_stationarity=False,
            enforce_invertibility=False
        ).fit(disp=False)
        pred = ajuste.forecast(steps=test_size, exog=test[exog_cols].astype(float).values)
        mets = metricas_regresion(test[target_col].values, pred)
        mets["modelo"] = "SARIMAX"
        pred_df = pd.DataFrame({"fecha_mes": fechas_test, "y_real": test[target_col].values, "y_pred": pred})
        return mets, pred_df, None
    except Exception as e:
        return None, None, str(e)


def forecast_sarimax(df_ts, target_col, exog_cols):
    if len(exog_cols) == 0:
        return None
    d, *_ = _train_test_ts(df_ts, target_col, extra_cols=exog_cols)
    if d is None:
        return None
    try:
        ajuste = SARIMAX(
            d[target_col].astype(float).values,
            exog=d[exog_cols].astype(float).values,
            order=(1, 1, 1),
            enforce_stationarity=False,
            enforce_invertibility=False
        ).fit(disp=False)
        ult_exog = pd.concat([d[exog_cols].iloc[-1:]] * HORIZONTE_MESES, ignore_index=True)
        pred = ajuste.forecast(steps=HORIZONTE_MESES, exog=ult_exog.astype(float).values)
        fechas_futuras = pd.date_range(start=d["fecha_mes"].max() + pd.offsets.MonthBegin(1), periods=HORIZONTE_MESES, freq="MS")
        return pd.DataFrame({"fecha_mes": fechas_futuras, "forecast": pred})
    except Exception:
        return None
```

```python
def parsear_fecha_robusta(serie):
    """
    Convierte fechas almacenadas como enteros o texto en distintos formatos observados en los datos.
    """
    s = pd.Series(serie, index=serie.index)
    out = pd.Series(pd.NaT, index=serie.index, dtype="datetime64[ns]")

    try:
        direct = pd.to_datetime(s, errors="coerce", dayfirst=True, format="mixed")
    except Exception:
        direct = pd.to_datetime(s, errors="coerce", dayfirst=True)

    out.loc[direct.notna()] = direct.loc[direct.notna()]

    mask = out.isna()
    if mask.any():
        parsed_individual = s.loc[mask].apply(lambda x: pd.to_datetime(x, errors="coerce", dayfirst=True))
        out.loc[parsed_individual.notna().index] = out.loc[parsed_individual.notna().index]
        idx_ok = parsed_individual.notna()
        if idx_ok.any():
            out.loc[parsed_individual.index[idx_ok]] = parsed_individual.loc[idx_ok].values

    mask = out.isna()
    if mask.any():
        s_txt = s.loc[mask].astype(str).str.strip()
        s_num = s_txt.str.replace(r"[^0-9]", "", regex=True)

        excel_mask = s_num.str.fullmatch(r"\d{5}")
        if excel_mask.any():
            excel_vals = pd.to_numeric(s_num.loc[excel_mask], errors="coerce")
            excel_dates = pd.to_datetime(excel_vals, unit="D", origin="1899-12-30", errors="coerce")
            out.loc[excel_dates.index] = excel_dates.values

        mask = out.isna()
        if mask.any():
            s_num = s.loc[mask].astype(str).str.strip().str.replace(r"[^0-9]", "", regex=True)
            s_num = s_num.where(~s_num.str.len().eq(7), "0" + s_num)

            for fmt in ["%d%m%Y", "%Y%m%d", "%d%m%y", "%y%m%d"]:
                parsed = pd.to_datetime(s_num, format=fmt, errors="coerce")
                idx_ok = parsed.notna()
                if idx_ok.any():
                    out.loc[parsed.index[idx_ok]] = parsed.loc[idx_ok].values

    return out
```

## BLOQUE 14 — Modelado: preparación de datos

```python

resultados_metricas = []
resultados_predicciones = {}
resultados_forecast = {}

BASE_MODELADO = monthly_total_model.copy() if "monthly_total_model" in globals() else monthly_total.copy() if "monthly_total" in globals() else None
EXOG_COLS = EXOG_COLS if "EXOG_COLS" in globals() else []

if BASE_MODELADO is not None:
    print("Iniciando modelado...")
    print("Base de modelado:", "monthly_total_model" if "monthly_total_model" in globals() else "monthly_total")
    print("Variables externas usadas:", EXOG_COLS)
else:
    print("[ERROR] No existe una base mensual para modelar.")
```

## BLOQUE 15 — Modelado por variable objetivo

```python

if BASE_MODELADO is not None:
    for target_col in VARIABLES_OBJETIVO:
        print("=" * 80)
        print(f"VARIABLE OBJETIVO: {target_col}")
        print("=" * 80)

        df_feat = crear_features_temporales_ext(BASE_MODELADO, target_col, EXOG_COLS)

        # 1) ARIMA
        mets, pred_df, err = entrenar_y_evaluar_arima(BASE_MODELADO, target_col)
        if mets is not None:
            resultados_metricas.append({"variable": target_col, **mets})
            resultados_predicciones[f"{target_col}_ARIMA"] = pred_df
            forecast_df = forecast_arima(BASE_MODELADO, target_col)
            if forecast_df is not None:
                forecast_df["modelo"] = "ARIMA"
                forecast_df["variable"] = target_col
                resultados_forecast[f"{target_col}_ARIMA"] = forecast_df
            print("[OK] ARIMA evaluado.")
        else:
            print(f"[WARNING] ARIMA no disponible para {target_col}: {err}")

        # 2) Exponential Smoothing
        mets, pred_df, err = entrenar_y_evaluar_exp_smoothing(BASE_MODELADO, target_col)
        if mets is not None:
            resultados_metricas.append({"variable": target_col, **mets})
            resultados_predicciones[f"{target_col}_ExpSmoothing"] = pred_df
            forecast_df = forecast_exp_smoothing(BASE_MODELADO, target_col)
            if forecast_df is not None:
                forecast_df["modelo"] = "Exponential Smoothing"
                forecast_df["variable"] = target_col
                resultados_forecast[f"{target_col}_ExpSmoothing"] = forecast_df
            print("[OK] Exponential Smoothing evaluado.")
        else:
            print(f"[WARNING] Exponential Smoothing no disponible para {target_col}: {err}")

        # 3) SARIMAX con exógenas BCCH
        if len(EXOG_COLS) > 0:
            mets, pred_df, err = entrenar_y_evaluar_sarimax(BASE_MODELADO, target_col, EXOG_COLS)
            if mets is not None:
                resultados_metricas.append({"variable": target_col, **mets})
                resultados_predicciones[f"{target_col}_SARIMAX"] = pred_df
                forecast_df = forecast_sarimax(BASE_MODELADO, target_col, EXOG_COLS)
                if forecast_df is not None:
                    forecast_df["modelo"] = "SARIMAX"
                    forecast_df["variable"] = target_col
                    resultados_forecast[f"{target_col}_SARIMAX"] = forecast_df
                print("[OK] SARIMAX evaluado.")
            else:
                print(f"[WARNING] SARIMAX no disponible para {target_col}: {err}")

        # 4) Regresión Lineal
        mets, pred_df, err = entrenar_y_evaluar_supervisado_ext(
            df_feat, target_col, "Regresión Lineal", LinearRegression(), EXOG_COLS
        )
        if mets is not None:
            resultados_metricas.append({"variable": target_col, **mets})
            resultados_predicciones[f"{target_col}_RegresionLineal"] = pred_df
            forecast_df = forecast_supervisado_ext(df_feat, target_col, LinearRegression(), EXOG_COLS)
            if forecast_df is not None:
                forecast_df["modelo"] = "Regresión Lineal"
                forecast_df["variable"] = target_col
                resultados_forecast[f"{target_col}_RegresionLineal"] = forecast_df
            print("[OK] Regresión Lineal evaluada.")
        else:
            print(f"[WARNING] Regresión Lineal no disponible para {target_col}: {err}")

        # 5) Random Forest
        mets, pred_df, err = entrenar_y_evaluar_supervisado_ext(
            df_feat, target_col, "Random Forest",
            RandomForestRegressor(n_estimators=300, random_state=42, max_depth=6),
            EXOG_COLS
        )
        if mets is not None:
            resultados_metricas.append({"variable": target_col, **mets})
            resultados_predicciones[f"{target_col}_RandomForest"] = pred_df
            forecast_df = forecast_supervisado_ext(
                df_feat, target_col,
                RandomForestRegressor(n_estimators=300, random_state=42, max_depth=6),
                EXOG_COLS
            )
            if forecast_df is not None:
                forecast_df["modelo"] = "Random Forest"
                forecast_df["variable"] = target_col
                resultados_forecast[f"{target_col}_RandomForest"] = forecast_df
            print("[OK] Random Forest evaluado.")
        else:
            print(f"[WARNING] Random Forest no disponible para {target_col}: {err}")

        # 6) XGBoost
        if xgboost_disponible:
            mets, pred_df, err = entrenar_y_evaluar_supervisado_ext(
                df_feat, target_col, "XGBoost",
                XGBRegressor(n_estimators=200, learning_rate=0.05, max_depth=4, random_state=42),
                EXOG_COLS
            )
            if mets is not None:
                resultados_metricas.append({"variable": target_col, **mets})
                resultados_predicciones[f"{target_col}_XGBoost"] = pred_df
                forecast_df = forecast_supervisado_ext(
                    df_feat, target_col,
                    XGBRegressor(n_estimators=200, learning_rate=0.05, max_depth=4, random_state=42),
                    EXOG_COLS
                )
                if forecast_df is not None:
                    forecast_df["modelo"] = "XGBoost"
                    forecast_df["variable"] = target_col
                    resultados_forecast[f"{target_col}_XGBoost"] = forecast_df
                print("[OK] XGBoost evaluado.")
            else:
                print(f"[WARNING] XGBoost no disponible para {target_col}: {err}")

        # 7) LightGBM
        if lightgbm_disponible:
            mets, pred_df, err = entrenar_y_evaluar_supervisado_ext(
                df_feat, target_col, "LightGBM",
                LGBMRegressor(n_estimators=200, learning_rate=0.05, max_depth=4, random_state=42, verbose=-1),
                EXOG_COLS
            )
            if mets is not None:
                resultados_metricas.append({"variable": target_col, **mets})
                resultados_predicciones[f"{target_col}_LightGBM"] = pred_df
                forecast_df = forecast_supervisado_ext(
                    df_feat, target_col,
                    LGBMRegressor(n_estimators=200, learning_rate=0.05, max_depth=4, random_state=42, verbose=-1),
                    EXOG_COLS
                )
                if forecast_df is not None:
                    forecast_df["modelo"] = "LightGBM"
                    forecast_df["variable"] = target_col
                    resultados_forecast[f"{target_col}_LightGBM"] = forecast_df
                print("[OK] LightGBM evaluado.")
            else:
                print(f"[WARNING] LightGBM no disponible para {target_col}: {err}")
else:
    print("[ERROR] No existe base de modelado.")
```

## Importancia de variables en modelos basados en árboles

A continuación se estima la importancia relativa de las variables explicativas utilizando un modelo `Random Forest` para cada variable objetivo. Esta lectura permite identificar si los rezagos propios de la serie dominan la predicción o si las variables externas del Banco Central aportan señal adicional.

Esta visualización es especialmente útil para la discusión metodológica, ya que ayuda a responder si el componente exógeno realmente agrega capacidad explicativa al problema.


```python
# Importancia de variables con Random Forest
from sklearn.ensemble import RandomForestRegressor

base_imp = monthly_total_model.copy() if "monthly_total_model" in globals() else monthly_total.copy() if "monthly_total" in globals() else None
EXOG_COLS_LOCAL = EXOG_COLS if "EXOG_COLS" in globals() else []

if base_imp is not None:
    importancias_resumen = []

    for target_col in [c for c in VARIABLES_OBJETIVO if c in base_imp.columns]:
        df_feat_imp = crear_features_temporales_ext(base_imp, target_col, EXOG_COLS_LOCAL).dropna().copy()

        feature_cols = ["t", "mes", "anio", "lag_1", "lag_2", "lag_3", "rolling_3", "rolling_6"]
        for c in EXOG_COLS_LOCAL:
            for suff in ["_lag1", "_lag2"]:
                colx = f"{c}{suff}"
                if colx in df_feat_imp.columns:
                    feature_cols.append(colx)

        if len(df_feat_imp) < 12:
            print(f"[INFO] Datos insuficientes para calcular importancia en {target_col}.")
            continue

        X = df_feat_imp[feature_cols]
        y = df_feat_imp[target_col]

        rf_imp = RandomForestRegressor(n_estimators=300, random_state=42, max_depth=6)
        rf_imp.fit(X, y)

        imp_df = pd.DataFrame({
            "feature": feature_cols,
            "importance": rf_imp.feature_importances_
        }).sort_values("importance", ascending=False).reset_index(drop=True)

        imp_df["variable_objetivo"] = target_col
        importancias_resumen.append(imp_df)

        top_imp = imp_df.head(12).sort_values("importance", ascending=True)

        plt.figure(figsize=(10, 6))
        plt.barh(top_imp["feature"], top_imp["importance"])
        plt.title(f"Importancia de variables - Random Forest ({target_col})")
        plt.xlabel("Importancia")
        plt.ylabel("Variable")
        plt.tight_layout()
        guardar_figura(f"importancia_variables_rf_{target_col}.png")
        plt.show()

        display(imp_df.head(15))

    if len(importancias_resumen) > 0:
        importancias_df = pd.concat(importancias_resumen, ignore_index=True)
        guardar_tabla(importancias_df, "importancia_variables_random_forest.csv")
else:
    print("[WARNING] No existe base mensual para calcular importancia de variables.")
```

El análisis de residuos permite verificar si los errores del modelo se distribuyen de manera razonable en el tiempo o si existen sesgos, meses problemáticos o errores extremos concentrados. Esto es particularmente importante cuando se trabaja con series cortas o con posibles outliers.

- y residuo versus valor predicho.

Finalmente, se incorpora una descomposición temporal para cada variable objetivo. Esta técnica permite separar visualmente la serie en:
- **tendencia**,
- **estacionalidad**,
- **residuo**.

El objetivo es evaluar si existen patrones sistemáticos en el tiempo que justifiquen el uso de modelos clásicos de series temporales, como `ARIMA`, `SARIMAX` o `Exponential Smoothing`. Dado el tamaño relativamente acotado de la serie, esta lectura debe entenderse como exploratoria.


```python
# Descomposición temporal de las series objetivo
from statsmodels.tsa.seasonal import seasonal_decompose

base_decomp = monthly_total.copy() if "monthly_total" in globals() else None

if base_decomp is not None:
    for variable in [c for c in VARIABLES_OBJETIVO if c in base_decomp.columns]:
        serie = base_decomp[["fecha_mes", variable]].dropna().copy().sort_values("fecha_mes")
        serie = serie.set_index("fecha_mes")[variable]

        if len(serie) >= 24:
            period = 12
        elif len(serie) >= 12:
            period = 6
        else:
            print(f"[INFO] No hay suficientes datos para descomponer {variable}.")
            continue

        try:
            descomp = seasonal_decompose(serie, model="additive", period=period, extrapolate_trend='freq')
            fig = descomp.plot()
            fig.set_size_inches(12, 8)
            fig.suptitle(f"Descomposición temporal - {variable} (period={period})", y=1.02)
            plt.tight_layout()
            guardar_figura(f"descomposicion_temporal_{variable}.png")
            plt.show()
        except Exception as e:
            print(f"[WARNING] No se pudo descomponer {variable}: {e}")
else:
    print("[WARNING] No existe monthly_total para descomposición temporal.")
```

## Forecast exploratorio por categoría HS4

Como complemento, se genera un forecast simple por categoría utilizando un modelo parsimonioso. El objetivo no es reemplazar el forecast agregado del proyecto, sino identificar qué categorías muestran mayor estabilidad y cuáles requieren una lectura más cautelosa.

Cuando la longitud de la serie es suficiente, se utiliza suavizamiento exponencial, en caso contrario, se usa una proyección naive basada en el último valor observado.

```python
if "monthly_by_hs" in globals() and not monthly_by_hs.empty:
    forecast_hs_rows = []
    for hs in sorted(monthly_by_hs["hs4"].astype(str).unique()):
        temp_hs = monthly_by_hs[monthly_by_hs["hs4"].astype(str) == hs].sort_values("fecha_mes").copy()
        if temp_hs.empty:
            continue

        idx_futuro = pd.date_range(
            start=temp_hs["fecha_mes"].max() + pd.offsets.MonthBegin(1),
            periods=HORIZONTE_MESES,
            freq="MS"
        )

        for variable in VARIABLES_OBJETIVO:
            serie = temp_hs[["fecha_mes", variable]].dropna().copy()
            if len(serie) < 3:
                continue

            try:
                if len(serie) >= 12:
                    modelo_hs = ExponentialSmoothing(serie[variable].astype(float), trend="add", seasonal=None).fit()
                    pred_hs = modelo_hs.forecast(HORIZONTE_MESES)
                    metodo = "ExponentialSmoothing"
                else:
                    pred_hs = np.repeat(float(serie[variable].iloc[-1]), HORIZONTE_MESES)
                    metodo = "Naive"
            except Exception:
                pred_hs = np.repeat(float(serie[variable].iloc[-1]), HORIZONTE_MESES)
                metodo = "Naive"

            for fecha_f, valor_f in zip(idx_futuro, pred_hs):
                forecast_hs_rows.append({
                    "hs4": hs,
                    "variable_objetivo": variable,
                    "fecha_mes": fecha_f,
                    "forecast": float(valor_f),
                    "metodo": metodo
                })

    forecast_hs_df = pd.DataFrame(forecast_hs_rows)
    if not forecast_hs_df.empty:
        display(forecast_hs_df.head(20))
        guardar_tabla(forecast_hs_df, "forecast_exploratorio_por_hs4.csv")

        for variable in VARIABLES_OBJETIVO:
            plt.figure(figsize=(12, 6))
            for hs in sorted(monthly_by_hs["hs4"].astype(str).unique()):
                hist = monthly_by_hs[(monthly_by_hs["hs4"].astype(str) == hs)].sort_values("fecha_mes")
                fut = forecast_hs_df[(forecast_hs_df["hs4"].astype(str) == hs) & (forecast_hs_df["variable_objetivo"] == variable)].sort_values("fecha_mes")
                if not hist.empty:
                    plt.plot(hist["fecha_mes"], hist[variable], label=f"{hs} histórico")
                if not fut.empty:
                    plt.plot(fut["fecha_mes"], fut["forecast"], linestyle="--", label=f"{hs} forecast")
            plt.title(f"Forecast exploratorio por categoría HS4 - {variable}")
            plt.xlabel("Fecha")
            plt.ylabel(variable)
            plt.xticks(rotation=45)
            plt.legend(ncol=2, fontsize=8)
            plt.tight_layout()
            guardar_figura(f"forecast_exploratorio_hs4_{variable}.png")
            plt.show()
    else:
        print("[WARNING] No fue posible construir forecast exploratorio por categoría.")
else:
    print("[WARNING] No existe monthly_by_hs para forecast exploratorio por categoría.")
```

## Limitaciones del proyecto

Aun cuando el proyecto permite construir una solución analítica defendible y útil, es importante explicitar sus principales limitaciones.

### 1. Longitud de la serie temporal
La cantidad de períodos disponibles es reducida (**24 meses**), lo que limita la captura robusta de estacionalidad y ciclos más largos.

### 2. Variables exógenas aún acotadas
Si bien se incorporaron variables externas del Banco Central de Chile, el conjunto sigue siendo acotado y no necesariamente captura todos los determinantes de las importaciones.

### 3. Calidad y heterogeneidad de los datos
Al trabajar con múltiples archivos y estructuras, existe riesgo de inconsistencias de formato, diferencias de codificación y presencia de outliers. En particular, la variable **cantidad** presenta al menos un valor extremadamente alto que afecta la estabilidad del modelamiento.

### 4. Desempeño desigual entre variables
No todas las variables presentan el mismo nivel de predictibilidad. En especial, **cantidad** mostró métricas débiles y proyecciones negativas, por lo que debe interpretarse con mucha cautela.

### 5. Alcance limitado del modelo
El forecast obtenido no representa directamente demanda comercial ni decisiones reales de compra de la empresa, sino una señal construida a partir de importaciones agregadas observadas en los datos públicos.

## Recomendaciones y mejoras futuras

5. **Profundizar la segmentación**, avanzando hacia análisis por subcategorías comerciales o combinaciones más específicas de productos.
6. **Profundizar la comparación de modelos**, mediante ajuste fino de hiperparámetros, validación temporal más robusta y comparación con otros enfoques clásicos y avanzados de series temporales.

Estas mejoras permitirían avanzar desde una herramienta de apoyo exploratoria hacia una solución analítica de mayor precisión y utilidad operativa.
